```python
r_cols = ['user_id','movie_id','rating','unix_timestamp']
ratings = pd.read_csv('C:/Users/Fortu/Downloads/Aut2024/Info 376/ml-100k/ml-100k/u.data', sep='\t',names=r_cols,encoding='latin-1')
```

```python
ratings.shape
```

```python
# Drop the columns we don't need. axis=1 means the operation should be performed on the columns.
ratings = ratings.drop('unix_timestamp', axis=1)
movies = movies.drop('video release date', axis=1)
movies = movies.drop('IMDB URL', axis=1)
```

```python
# Step-2: Construct user-movie rating matrix by doing pivot (it's like a join operation in databases)
u_m = ratings.pivot(index='user_id',columns='movie_id',values='rating').fillna(0)
```

```python
u_m.shape
```

```python
u_m.head(5)
```

```python
# Step-3: normalize ratings.
data = u_m.values
ratings_mean = np.mean(data,axis=1)
normalizedRating = data - ratings_mean.reshape(-1,1)
```

```python
normalizedRating
```

```python
# Step-4: SVD (matrix factorization)
from scipy.sparse.linalg import svds
U, sigma, Vt = svds(normalizedRating, k=100)
```

```python
U.shape
```

```python
sigma.shape
```

```python
Vt.shape
```

```python
sigma
```

```python
# Step-5: get the diagonal entries of sigma (singular values)
sigma = np.diag(sigma)
```

```python
sigma
```

```python
sigma.shape
```

```python
# Step-6: reshape and renormalize ratings using weighted/important components/factors
all_user_ratings = np.dot(np.dot(U,sigma), Vt) + ratings_mean.reshape(-1,1)
```

```python
all_user_ratings
```

```python
# Step-7: construct a dataframe with all these user-movie ratings predictions.
preds = pd.DataFrame(all_user_ratings, columns=u_m.columns)
```

```python
preds.head(5)
```

```python
preds.shape
```

```python
# Given a matrix of predictions for movie ratings, produce recommendations for a given user
def recommend_movies(predictions_df,userId,movies_df,original_ratings_df,num_recs):
  # Get and sort the user's predictions
  user_row_number = userId - 1 # userId starts at 1, not 0
  sorted_user_predictions = predictions_df.iloc[user_row_number].sort_values(ascending=False)

  # Get the user's data and merge in the movie information
  user_data = original_ratings_df[original_ratings_df['user_id']==(userId)] # Row with user info
  user_full = (user_data.merge(movies_df, how='left',left_on='movie_id', right_on='movieid').sort_values(['rating'],ascending=False))
  print('User {0} has already rated {1} movies.'.format(userId,user_full.shape[0]))
  print('Recommending highest {0} predicted ratings movies not already rated.'.format(num_recs))

  # Recommend the highest predicted rating movies that the user hasn't seen yet.
  recommendations = (movies_df[~movies_df['movieid'].isin(user_full['movie_id'])]).merge(pd.DataFrame(sorted_user_predictions)
                                .reset_index(), how='left',left_on='movieid',right_on='movie_id').rename(
                                  columns={user_row_number:'Predictions'}).sort_values('Predictions',ascending=False).iloc[:num_recs,:-1]
  return user_full, recommendations
```

```python
# What we want to do
already_rated, recommendations = recommend_movies(preds,1,movies,ratings,10)
```
