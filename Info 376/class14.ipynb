```python
# Item-based collaborative filtering
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./songs_sample.csv')
```

```python
song_matrix = pd.pivot_table(df,values='listen_time',index='title',columns='userid')
```

```python
song_matrix.head()
```

```python
song_matrix = song_matrix.fillna(0)
```

```python
song_matrix.head()
```

```python
# Normalize the listen time by adjusting it around the mean for a user.
# axis=1 means the operation should be performed on the columns.
song_norm = song_matrix.apply(lambda x: x - np.nanmean(x), axis=1)
song_norm.head()
```

```python
# Compute item-item similarity
from sklearn.metrics.pairwise import cosine_similarity
import operator

item_sim_df = pd.DataFrame(cosine_similarity(song_norm,song_norm),index=song_norm.index,columns=song_norm.index)
```

```python
item_sim_df.head()
```

```python
# Given a song, find the similar songs
def get_similar_songs(songid):
  if songid not in song_norm.index:
    print("{} item not available".song_norm.index)
    return None, None

  else:
    sim_songs = item_sim_df.sort_values(by=songid, ascending=False).index[1:]
    sim_scores = item_sim_df.sort_values(by=songid, ascending=False).loc[:,songid].tolist()[1:]
    return sim_songs, sim_scores
```

```python
# What we would like to do
songs, scores = get_similar_songs("'Till_I_Collapse")
```

```python
for x, y in zip(songs[:10], scores[:10]):
  print("{} with similarity of {}".format(x,y))
```
