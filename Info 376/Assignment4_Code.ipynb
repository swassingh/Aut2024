1. Explain how language models work for IR in your own words. Why are they called 'language models'? Respond in at most half a page. [2 points]

- Language models in IR are essentially algorithms that use probability to predict the likelyhood of a document to answer the user's query. These algorithms are called language models because of how they were designed to represent or *model* patterns, structure, and relationships between words in a given language. They learn from large amounts of textual data, hence the language part of their name, to predict how language behaves.
